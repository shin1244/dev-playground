```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
"""
스태킹 연습(유방암 분류기)
RF, KNN, ADABOOSTING, DT 1차 prediction
LR(meta model) 2차 PREDICTION
"""
```

```python
cancer_data=load_breast_cancer()
cancer_data
```

```python
X_data=cancer_data.data
y_label=cancer_data.target
```

```python
X_data
y_label
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_data,y_label,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
knn_clf=KNeighborsClassifier(n_neighbors=4)
rf_clf=RandomForestClassifier(n_estimators=100,random_state=0)
dt_clf=DecisionTreeClassifier()
ada_clf=AdaBoostClassifier(n_estimators=100)
```

```python
lr_final_clf=LogisticRegression() #메타모델
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error
import warnings
```

```python
X_train.shape #(455, 30), 455 / k(7) => 1 fold 당 65개 data
y_train.shape#(455,)
X_test.shape #(114, 30)
```

```python
#1. 각 모델로 K(7)교차 검증 수행 -> 최종 메타모델의 학습 데이터와 테스트 데이터 생성
def get_stacking_base_datasets(model,X_train_n, y_train_n, X_test_n, n_folds):
    kf=KFold(n_splits=n_folds, shuffle=False )
    #메타 모델용 학습데이터를 생성하기 위한 배열 초기화
    train_fold_pred=np.zeros((X_train_n.shape[0],1)) #[455,1]
    test_pred=np.zeros((X_test_n.shape[0],n_folds)) #[114,7]
    for folder_counter,(train_index,valid_index) in enumerate(kf.split(X_train_n)):
        X_tr=X_train_n[train_index]
        y_tr=y_train_n[train_index]
        X_vl=X_train_n[valid_index]
        model.fit(X_tr, y_tr)
        #각 모델의 예측 결과로 생성한 학습 데이터(메타 모델의 학습데이터로 사용예정)
        train_fold_pred[valid_index,:]=model.predict(X_vl).reshape(-1,1)
        #각 모델의 예측 결과로 생성한 테스트 데이터(메타 모델의 테스트데이터로 사용예정)
        test_pred[:,folder_counter]=model.predict(X_test_n)  #test_pred=[114,7]  
    print(test_pred.shape)
    test_pred_mean=np.mean(test_pred,axis=1).reshape(-1,1) #test_pred=[114,7] =>(114, 1)
#     print(train_fold_pred.shape) #(455, 1)
#     print(test_pred_mean.shape) #(114, 1)
#     print("="*50)
    return train_fold_pred,test_pred_mean        
        
        
        
#                              print(folder_counter) #0~6
#                              print(b)
#                              print(c)
#                              print("="*50)
    
#     print(train_fold_pred.shape, test_pred.shape)
#     print("="*50)
#get_stacking_base_datasets(rf_clf,X_train,y_train,X_test,7)
knn_train,knn_test=get_stacking_base_datasets(knn_clf,X_train,y_train,X_test,7)
rf_train,rf_test=get_stacking_base_datasets(rf_clf,X_train,y_train,X_test,7)
dt_train,dt_test=get_stacking_base_datasets(dt_clf,X_train,y_train,X_test,7)
ada_train,ada_test=get_stacking_base_datasets(ada_clf,X_train,y_train,X_test,7)

```

```python
#2. 각 모델의 예측 결과들을 옆으로 쌓기(stacking)
Stack_final_X_train=np.concatenate((knn_train,rf_train,dt_train,ada_train),axis=1)
Stack_final_X_test=np.concatenate((knn_test,rf_test,dt_test,ada_test),axis=1)
```

```python
np.shape(Stack_final_X_train) #(455, 4) 알고리즘 4개에 대한 TRAIN 예측 결과를 stack
np.shape(Stack_final_X_test)#(114, 4)알고리즘 4개에 대한 test 예측 결과를 stack
```

```python
#3. 메타 모델 학습 -> 예측 -> 원본 테스트 데이터의 레이블과 비교 => 정확도 출력
lr_final_clf.fit(Stack_final_X_train,y_train)
```

```python
final_pred=lr_final_clf.predict(Stack_final_X_test)
```

```python
print("최종 스태킹 모델 정확도 : {0:.4f}".format(accuracy_score(final_pred, y_test)))
```
