```python
Data_set=np.loadtxt("data/ThoraricSurgery3.csv", delimiter=",")
```

```python
x=Data_set[:,0:16] #환자상태
y=Data_set[:,16] #생존여부, 0:생존, 1:사망
```

```python
#모델 구성
model=Sequential()
model.add(Dense(30,input_dim=16 , activation='relu' ))#0<=relu<=inf
model.add(Dense(1, activation='sigmoid'))#0<=sig<=1 => 0.5기준 => 1/0
```

```python
#모델 실행
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history=model.fit(x,y,epochs=5, batch_size=1)
```

```python
x = np.array([2,
              4,
              6,
              8]) #공부시간, => input_dim=1

#               시 출 결 조
#x = np.array([[2, 4, 6, 8],[2, 1, 1, 0],[1, 0, 0, 0]) =>input_dim=4

y = np.array([81, 93, 91, 97])  #점수
#선형회귀모델 -> 만약 7시간 공부를 했다면 점수?
```

```python
model = Sequential()
```

```python
model.add(Dense(1, input_dim=1 ,activation='linear'))
```

```python
model.compile(optimizer='sgd', loss='mse')
```

```python
model.fit(x,y,epochs=2000)
```

```python
model.predict([7])
```
