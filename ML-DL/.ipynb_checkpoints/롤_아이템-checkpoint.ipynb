```python
while True:
    if keyboard.is_pressed('tab'):
        scr = ImageGrab.grab()
        for x in range(5):
            itemcap1 = scr.crop((715, 330+76*x, 920, 380+76*x))
            itemcap2 = scr.crop((1287, 330+76*x, 1492, 380+76*x))
            ch_list1 = []
            ch_list2 = []
            for i in range(6):
                item_img1 = itemcap1.crop((32*i+2*i, 2, 32*(i+1)+2*i, 34))
                item_img2 = itemcap2.crop((32*i+2*i, 2, 32*(i+1)+2*i, 34))
                
                
                image1 = cv.cvtColor(np.array(item_img1), cv.COLOR_BGR2HSV)
                image2 = cv.cvtColor(np.array(item_img2), cv.COLOR_BGR2HSV)
                score_list1 = []
                score_list2 = []
                for j in item_icon:
                    result1 = cv.matchTemplate(image1, j, cv.TM_CCOEFF_NORMED )
                    result2 = cv.matchTemplate(image2, j, cv.TM_CCOEFF_NORMED )
                    min_val1, max_val1, min_loc1, max_loc1 = cv.minMaxLoc(result1)
                    min_val2, max_val2, min_loc2, max_loc2 = cv.minMaxLoc(result2)
                    score_list1.append(max_val1)
                    score_list2.append(max_val2)
                if max(score_list1) > 0.5:
                    ch_list1.append(item_cost[np.argmax(score_list1)])
                if max(score_list2) > 0.5:
                    ch_list2.append(item_cost[np.argmax(score_list2)])
            print(x + 1)
            print(ch_list1)
            print(ch_list2)
            print(x + 1)
    elif keyboard.is_pressed('escape'):
        break
    time.sleep(3)
```
