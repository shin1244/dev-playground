```python
from glob import glob

img_list = glob('dataset/train/images/*.jpg')

print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.1)

print(len(train_img_list), len(val_img_list))
```

```python
with open('dataset/train.txt', 'w') as f:
    f.write('\n'.join(train_img_list) + '\n')

with open('dataset/valid.txt', 'w') as f:
    f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml

with open('dataset/data.yaml', 'r', encoding='utf-8') as f:
    data = yaml.safe_load(f)

print(data)

data['train'] = '../dataset/train.txt'
data['val'] = '../dataset/valid.txt'
data['names'] = ['블랑 맥주',
                 '버드와이저 맥주',
                 '처음처럼 소주',
                 '카스0.0 맥주',
                 '카스 맥주',
                 '청하 소주',
                 '참이슬오리지널 소주',
                 '참이슬 소주',
                 '칭따오 맥주',
                 '클라우드 맥주',
                 '이슬톡톡 과실주',
                 '기네스 맥주',
                 '좋은데이 소주',
                 '하이네켄 맥주',
                 '호가든 맥주',
                 '진로 소주',
                 '산사춘 소주',
                 '스텔라아르투아 맥주',
                 '테라 맥주']
with open('dataset/data.yaml', 'w', encoding="utf-8") as f:
    yaml.dump(data, f, allow_unicode=True)

print(data)
```
