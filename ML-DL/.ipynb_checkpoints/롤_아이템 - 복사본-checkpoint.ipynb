```python
from PIL import ImageGrab
from PIL import Image
import keyboard
import cv2 as cv
import numpy as np
import time
```

```python
item_icon = []
for i in range(len(item_cost)):
    if item_cost[i] != 0:
        item_icon.append(cv.imread(f'item/{i}img.png'))
item_cost_f = list(filter(lambda x: x != 0, item_cost))

item_icon = [cv.resize(i, (32, 32)) for i in item_icon]
item_icon = [cv.cvtColor(i, cv.COLOR_BGR2HSV) for i in item_icon]
```

```python
while True:
    if keyboard.is_pressed('='):
        scr = ImageGrab.grab()
        for x in range(5):
            itemcap1 = scr.crop((715, 330+76*x, 920, 380+76*x))
            itemcap2 = scr.crop((1287, 330+76*x, 1492, 380+76*x))
            ch_list1 = []
            ch_list2 = []
            for i in range(6):
                item_img1 = itemcap1.crop((32*i+2*i, 2, 32*(i+1)+2*i, 34))
                item_img2 = itemcap2.crop((32*i+2*i, 2, 32*(i+1)+2*i, 34))
                
                
                image1 = cv.cvtColor(np.array(item_img1), cv.COLOR_BGR2HSV)
                image2 = cv.cvtColor(np.array(item_img2), cv.COLOR_BGR2HSV)
                score_list1 = []
                score_list2 = []
                for j in item_icon:
                    result1 = cv.matchTemplate(image1, j, cv.TM_CCOEFF_NORMED )
                    result2 = cv.matchTemplate(image2, j, cv.TM_CCOEFF_NORMED )
                    min_val1, max_val1, min_loc1, max_loc1 = cv.minMaxLoc(result1)
                    min_val2, max_val2, min_loc2, max_loc2 = cv.minMaxLoc(result2)
                    score_list1.append(max_val1)
                    score_list2.append(max_val2)
                if max(score_list1) > 0.4:
                    ch_list1.append(item_cost[np.argmax(score_list1)])
                if max(score_list2) > 0.4:
                    ch_list2.append(item_cost[np.argmax(score_list2)])
            print(x + 1)
            print(ch_list1)
            print(ch_list2)
            print(x + 1)
    elif keyboard.is_pressed('escape'):
        break
    time.sleep(1)
```

```python
res = ur.urlopen('https://leagueoflegends.fandom.com/wiki/Item_(League_of_Legends)')
data = res.read()

soup = BeautifulSoup(data, 'html.parser')

item_click = soup.select('.tlist')
item_click = item_click[:-4]

item_links = []
for item in item_click:
    links = item.select('ul > li > div > div > a')
    item_links.extend(links)

item_click_list = []
for link in item_links:
    item_click_list.append(link['href'])
```

```python
item_cost = []

for idx,i in enumerate(item_click_list):
    res = ur.urlopen('https://leagueoflegends.fandom.com/' + i)
    data = res.read()

    soup = BeautifulSoup(data, 'html.parser')
    
    item_cost.append(soup.select_one('td:nth-child(1) > span > span:nth-child(2)'))
    item_img = soup.select_one('tr > td > div > img')
    item_url = item_img['src']
    ulib.urlretrieve(item_url, f'item/{idx}img.png')
    
item_cost = [0 if x is None else int(x.string) for x in item_cost]
```
