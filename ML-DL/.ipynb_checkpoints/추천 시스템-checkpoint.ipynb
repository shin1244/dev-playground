```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Eggs', 'Yogurt'],
           ['Onion', 'Nutmeg', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Ice cream', 'Eggs']]
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(dataset).transform(dataset)
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
freq = apriori(df, min_support=0.3, use_colnames=True)
```

```python
freq
```

```python
association_rules(freq, metric='lift', min_threshold=2)
```

```python
df = pd.read_csv('Groceries_dataset.csv', parse_dates=['Date'])
```

```python
df = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(lambda x: ', '.join(x)).reset_index()
```

```python
df
```

```python
it = []
for i in df['itemDescription']:
    it.append(i.split(', '))
```

```python
te = TransactionEncoder()
```

```python
len(te.columns_)
```

```python
item_df = pd.DataFrame(te.fit_transform(it), columns=te.columns_)
freq = apriori(item_df, min_support=0.001, use_colnames=True)
freq
```

```python
association_rules(freq, metric='lift', min_threshold=1)
```

```python
store_df = pd.read_csv('store_data.csv', header=None)
```

```python
store_df
```

```python
records=[]
for i in range(len(store_df)):
    records.append([str(store_df.values[i,j]) for j in range(len(store_df.columns)) if not pd.isna(store_df.values[i,j])])
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit_transform(records)
```

```python
te_ary.shape
te.columns_
df=pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
freq = apriori(df, min_support=0.01, use_colnames=True)
```

```python
freq
```

```python
association_rules(freq, metric='lift', min_threshold=2)
```
