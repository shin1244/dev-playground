```python
import os
import cv2 as cv
```

```python
cascade = cv.CascadeClassifier('C:/Users/jg999/Desktop/Drunk.v5-zibs.voc/raw.githubusercontent.com_opencv_opencv_master_data_haarcascades_haarcascade_frontalface_default.xml')
```

```python
def face_found(img, num):
  gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

  faces = cascade.detectMultiScale(gray, 1.1, 6, minSize=(64, 64))
  if len(faces) == 0:
    print("얼굴을 찾을 수 없습니다.")
  else:
    for idx, (x, y, width, height) in enumerate(faces):
      face_img = img[y:y+height, x:x+width]
      filename = f"face_{num, idx}.jpg"
      save_path = os.path.join("C:/Users/jg999/Desktop/Drunk.v5-zibs.voc/new", filename)
      cv.imwrite(save_path, face_img)
      print(f"{filename}로 얼굴 이미지 저장 완료")
```

```python
cham = 'C:/Users/jg999/Desktop/Drunk.v5-zibs.voc/train1'
cham_list = os.listdir(cham)
for idx, i in enumerate(cham_list):
  img_path = os.path.join(cham, i)
  image = cv.imread(img_path)
  face_found(image, idx)
```

```python
import os
import random
import shutil

source_folder = 'C:/Users/jg999/Desktop/crop_part1'  # 이미지가 저장된 폴더 경로
target_folder = 'C:/Users/jg999/Desktop/crop_part'  # 선택된 이미지를 저장할 폴더 경로
num_images = 1600  # 선택할 이미지의 개수

# 폴더가 없는 경우 생성
if not os.path.exists(target_folder):
    os.makedirs(target_folder)

# 폴더 내의 이미지 파일 리스트 가져오기
image_files = os.listdir(source_folder)

# 이미지 파일 리스트를 무작위로 섞기
random.shuffle(image_files)

# 지정된 개수만큼 이미지 선택 및 복사
for i in range(num_images):
    image_file = image_files[i]
    source_path = os.path.join(source_folder, image_file)
    target_path = os.path.join(target_folder, image_file)
    shutil.copyfile(source_path, target_path)

print(f"{num_images}개의 이미지를 성공적으로 저장했습니다.")
```
