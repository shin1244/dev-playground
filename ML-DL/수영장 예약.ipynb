```python
def now_status(url):
    try:
        res = ur.urlopen(url)
        data = res.read()
    except:
        return "오류"

    soup = BeautifulSoup(data, 'html.parser')

    status = soup.select('#cont > div > div.receipt_status_area > table > tbody > tr > td:nth-child(3)')
    return status[0].text
```
