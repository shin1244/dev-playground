```python
from pydub import AudioSegment
import librosa
from pydub.playback import play
import os
from PIL import Image
from random import choice
from random import randint
import numpy as np
import pandas as pd
```

```python
sound_file_list = os.listdir('D:/sound')
siren_file_list = os.listdir('D:/sound/siren')


result_list = []
for idx, i in enumerate(sound_file_list):
    song = AudioSegment.from_file(f"D:/sound/{i}")
    siren_choice = choice(siren_file_list)
    siren = AudioSegment.from_file(f"D:/sound/siren/{siren_choice}", format="mp3")
    siren = siren[9000:]
    start_time = randint(0, int(song.duration_seconds * 1000) - 5000)
    result = song.overlay(siren, position=start_time)
    result.export(f"D:/overlaysound/overlay{idx}.wav", format="wav")
    result_list.append([f'overlay{idx}.wav', siren_choice[:2], start_time])
```

```python
import os
import soundfile as sf
import librosa

siran_file_list = os.listdir('train_wav')

for i in siran_file_list:
    input_file = f'train_wav/{i}'
    output_file = f'train_10s/{i}'
    start_time = 10.0

    y, sr = librosa.load(input_file, sr=None)
    start_frame = int(start_time * sr)
    trimmed_y = y[start_frame:]

    sf.write(output_file, trimmed_y, sr)
```

```python
sound_file_list = os.listdir('D:/sound')
siran_file_list = os.listdir('train_10s')

result_list = []
for idx, i in enumerate(sound_file_list):
    song = AudioSegment.from_file(f"D:/sound/{i}")
    siren_choice = choice(siran_file_list)
    siren = AudioSegment.from_file(f"train_10s/{siren_choice}")
    all_time = song.duration_seconds
    siren_time = siren.duration_seconds
    start_time = randint(0, int(all_time * 1000) - 5000)
    result = song.overlay(siren, position=start_time)
    result.export(f"D:/overlay/overlay{idx}.wav", format="wav")
    end_time = start_time + int(siren_time * 1000) if start_time + int(siren_time * 1000) <= int(all_time * 1000) else int(all_time * 1000)
    result_list.append([f'overlay{idx}.wav', siren_choice[:2], start_time, end_time, int(all_time*1000)])
```

```python
pd.DataFrame(result_list, columns=['title', 'type', 'start_time', 'end_time', 'all_time']).to_csv('sound_csv.csv', encoding='utf-8-sig', index=False)
```
