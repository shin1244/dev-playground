```python
import random
import pandas as pd
import numpy as np
import os
import librosa
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

from tqdm.auto import tqdm

import warnings
warnings.filterwarnings(action='ignore') 
```

```python
train = pd.read_csv('sound/train.csv')
test = pd.read_csv('sound/test.csv')
sample_submission = pd.read_csv('sound/sample_submission.csv')
```

```python
Sample_list = []
for i in range(len(train)):
    y, sr = librosa.load('sound/train/'+train['SAMPLE_ID'][i]+'.wav')
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
    mfcc = librosa.feature.mfcc(S=librosa.power_to_db(mel_spec))
    mfcc_list = []
    for e in mfcc:
        mfcc_list.append(np.mean(e))
    Sample_list.append(mfcc_list)
```

```python
Sample_list = []
for i in range(len(test)):
    y, sr = librosa.load('sound/test/'+test['SAMPLE_ID'][i]+'.wav')
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
    mfcc = librosa.feature.mfcc(S=librosa.power_to_db(mel_spec))
    mfcc_list = []
    for e in mfcc:
        mfcc_list.append(np.mean(e))
    Sample_list.append(mfcc_list)
```

```python
lf = IsolationForest()
```

```python
lf.fit(Sample_list)
```

```python
result = lf.predict(Sample_list)
result
```

```python
result = np.where(result == 1, 0, result)
result = np.where(result == -1, 1, result)
```

```python
result
```

```python
sample_submission['LABEL'] = result
```

```python
sample_submission.to_csv('result.csv', index=False)
```
