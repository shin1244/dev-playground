```python
def is_japanese(text):
    pattern = r'[^\u3040-\u309F\u30A0-\u30FF\u4E00-\u9FFF\u31F0-\u31FF\uFF66-\uFF9F]'
    
    cleaned_text = re.sub(pattern, '', text)
    
    return bool(cleaned_text) and not re.search(pattern, cleaned_text)
```
