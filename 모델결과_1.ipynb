```python
from keras.models import load_model
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import sys
import librosa
from librosa import load
warnings.filterwarnings('ignore')
import shutil

from PIL import Image

loaded_model = load_model('cnn_model.hdf5')
import os
from tqdm import tqdm
```

```python
wav_folder = os.listdir('D:/test')
wav_folder = sorted(wav_folder, key=lambda filename: int(''.join(filter(str.isdigit, filename))))
real_result_list = []

for wav_file in tqdm(wav_folder):

    y, sr = load(f'D:/test/{wav_file}')

    time = int(3 * sr)
    len_s = len(y) // time
    audio_1s = np.array_split(y[:len_s * time], len_s)

    mel_list = []
    for i in audio_1s:
        mel_spec = librosa.feature.melspectrogram(y=i, sr=sr)
        S_db = librosa.power_to_db(mel_spec, ref=np.max)
        mel_list.append(S_db)

    result = {}

    for idx, i in enumerate(mel_list):
        librosa.display.specshow(i)
        plt.axis('off')  # 이미지에 축 제거
        plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
        plt.savefig(f'D:/siren/temp.png', dpi=128)
        plt.close()

        img = Image.open('D:/siren/temp.png')
        img = img.convert('RGB')
        img = img.resize((128, 128))
        img = np.array(img)
        img = img / 255.0
        img = np.expand_dims(img, axis=0)
        result[f'{idx + 1:03}.png'] = np.argmax(loaded_model.predict(img))

    os.remove('D:/siren/temp.png')

    result_list = list(result.values())
    print(result_list)

    c = 0
    count = []
    start = 0
    
    for i in range(len(result_list)):
        if c == 0 and result_list[i] != 0:
            c = 1
            count.append(result_list[i])
            start = i
        elif c == 1 and result_list[i] != 0:
            count.append(result_list[i])
        elif c == 1 and result_list[i] == 0:
            break
    if len(count) == 0:
        real_result_list.append([wav_file, 0, 0, 0])
    
    else:
        count_arr = np.array(count)
        real_result_list.append([wav_file, np.argmax(np.bincount(count_arr)), start * 3, start * 3 + len(count) * 3])
```

```python
df1 = pd.DataFrame(real_result_list, columns=['title', 'type', 'start_time', 'end_time'])
```
