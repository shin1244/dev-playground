```python
import pandas as pd
import numpy as np
import librosa
import os
from sklearn.ensemble import IsolationForest

import warnings
warnings.filterwarnings(action='ignore') 
```

```python
# 이용방법
# 1. "다람쥐 헌 쳇바퀴에 타고파"를 녹음한 5~10개의 파일을 트레이닝 음성 폴더에 넣는다.
# 2. 테스트 음성을 녹음하여 테스트 음성 폴더에 넣는다.
#테스트 음성 예시) "다람쥐 헌 쳇바퀴에 타고파"를 혀를 꼬아 녹음한 음성과 평범하게 녹음한 음성
```

```python
voice = '트레이닝 음성 폴더 경로'
voice_list = os.listdir(voice)
Sample_list = []
for i in voice_list:
    voice_path = os.path.join(voice, i)
    y, sr = librosa.load(voice_path)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
    mfcc = librosa.feature.mfcc(S=librosa.power_to_db(mel_spec))
    mfcc_list = []
    for e in mfcc:
        mfcc_list.append(np.mean(e))
    Sample_list.append(mfcc_list)
    
# 학습
lf = IsolationForest(n_estimators=50, max_samples=5, random_state=230609)
lf.fit(Sample_list)
```

```python
voice_test = '테스트 음성 폴더 경로'
voice_test_list = os.listdir(voice_test)
Sample_test_list = []
for i in voice_test_list:
    voice_path = os.path.join(voice_test, i)
    y, sr = librosa.load(voice_path)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
    mfcc = librosa.feature.mfcc(S=librosa.power_to_db(mel_spec))
    mfcc_list = []
    for e in mfcc:
        mfcc_list.append(np.mean(e))
    Sample_test_list.append(mfcc_list)
    
# 결과 1: 정상, -1: 비정상
lf.predict(Sample_test_list)
```
