# 단순 퓨리에 변환

```python
fft = np.fft.fft(y)
spectrum =np.abs(fft)
f = np.linspace(0,sr,len(spectrum))
```

```python
left_spectrum = spectrum[:int(len(spectrum)/2)]
left_f = f[:int(len(spectrum)/2)]
```

```python
plt.figure(figsize=fig_size)
plt.plot(left_f, left_spectrum)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.title("Power spectrum")
```

 - 푸리에 변환 결과는 복소공간에 형성되어 실수부와 허수부로 나뉘게 된다. 이 값을 절댓값을 취해 spectrum 사용한다.
 - Frequency는 초당 주기의 반복 빈도 수 이므로, 초당 샘플갯수인 sr보다 클 수 없다. 따라서 0 ~ frequency 사이에서 len(spectrum) 만큼 frequency domain을 만들어 주자.

# STFT → Spectrogram

```python
FIG_SIZE = (15,10)
hop_length = 512  # 전체 frame 수
n_fft = 2048  # frame 하나당 sample 수

# calculate duration hop length and window in seconds
hop_length_duration = float(hop_length)/sr
n_fft_duration = float(n_fft)/sr

# STFT
stft = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)

# 복소공간 값 절댓값 취하기
magnitude = np.abs(stft)

# magnitude > Decibels 
log_spectrogram = librosa.amplitude_to_db(magnitude)

# display spectrogram
plt.figure(figsize=FIG_SIZE)
librosa.display.specshow(log_spectrogram, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram (dB)")
```

# 푸리에변환 :
해당 신호를 주파수 영역으로 보았을때  어떤 주파수 성분이 얼마만큼 가지고 있는가를 가시적으로 표현
하지만, 어느 시간대에 주파수가 어떻게 변했는지 알 수가 없다.

```python
hop_length = 512  # 전체 frame 수
n_fft = 2048  # frame 하나당 sample 수 - sample갯수는  n_fft/2 +1 값
n_mels = 128

for i in range(len(machine_train)):
    if machine_train['FAN_TYPE'][i]==0:
        audio_path_0 = machine_train['SAMPLE_ID'][i]+'.wav'
        y_0, sr_0 = librosa.load('sound/train/'+audio_path_0)
        # FFT (Fast Fourier Transform)
        # y파형을 주파수 분석을 통해, 특정 시간에 주파수 성분이 어떻게 구성되어 있는지 확인할 수 있다.
        
#         fft = np.fft.fft(y_0) # fft 변환
#         spectrum =np.abs(fft)
#         f = np.linspace(0,sr,len(spectrum))
        #print(f)
        
        #  stft 변환
        stft = librosa.stft(f, n_fft = n_fft, hop_length = hop_length)
        spectrogram = np.abs(stft) # stst 변환
        #print("Spectogram :\n", spectrogram)

        # Mel Spectrogram 생성
        mel_spec = librosa.feature.melspectrogram(S=spectrogram, sr=sr, n_mels=n_mels, hop_length=hop_length)
        
        # dB 스케일 변환
        S_dB = librosa.power_to_db(mel_spec, ref=np.max) 
        
        # MFCC
        mfcc = librosa.feature.mfcc(S=S_dB, sr=sr)
        
        # np.fft.fft        
        # win_length는 FFT를 할 때 참조할 그래프의 길이
        # hop_length는 얼마만큼 시간 주기를 이동하면서 분석을 할 것인지에 대한 파라미터
        # n_fft는 win_length보다 길 경우 모두 zero padding해서 처리하기 위한 파라미터
        
        
        # STFT(Short-Time Fourier Transform)
        # 시간 정보가 유실되는 것을 방지하기 위해, 전에 정의한 시간의 간격(window 또는 frame) 
        # 단위로 쪼개어 푸리에 변환을 적용하는 기법
        # - window의 크기(n_fft)와 window 간에 겹치는 사이즈(hop_length)를 설정한다. 
        # 일반적으로는 n_fft의 1/4 정도가 겹치도록 설정한다고 한다.

        
        # librosa.power_to_db
        # 즉 주파수(frequency)를 가지는 신호의 성질은 dB 스케일에 정량적으로 비례하는 특성을 가지고 있다는 점입니다.
        # 따라서 주파수(frequency)는 magnitude(진폭)보다 그것의 지수를 취한 log 스케일에 비례하는 특성을 가지며, 
        # 그것을 개념적으로 쉽게 표현하기 위해 dB 스케일을 사용합니다.
        
        # melspectrogram
        #실제 주파수 정보를 인간의 청각 구조를 반영하여 수학적으로 변환하기 위한 대표적인 방법입니다
        
        
        
        
        
#     elif machine_train['FAN_TYPE'][i]==2:
#         audio_path_2 = machine_train['SAMPLE_ID'][i]+'.wav'
#         y_2, sr_2 = librosa.load('machine/train/'+audio_path_2)
#         print(audio_path_2)
#print(len(mfcc))
print(len(mfcc))
```

```python
hop_length = 512  # 전체 frame 수
n_fft = 2048  # frame 하나당 sample 수 - sample갯수는  n_fft/2 +1 값
n_mels = 128
mfcc_0_list = []
mfcc_2_list = []
for i in range(len(machine_train)):
    
    if machine_train['FAN_TYPE'][i]==0:
        audio_path_0 = machine_train['SAMPLE_ID'][i]+'.wav'
        y_0, sr_0 = librosa.load('sound/train/'+audio_path_0)
        stft_0 = librosa.stft(y_0, n_fft = n_fft, hop_length = hop_length)
        spectrogram_0 = np.abs(stft_0) 
        mel_spec_0 = librosa.feature.melspectrogram(S=spectrogram_0, sr=sr_0, n_mels=n_mels, hop_length=hop_length)
        S_dB_0 = librosa.power_to_db(mel_spec_0, ref=np.max)
        mfcc_0 = librosa.feature.mfcc(S=S_dB_0, sr=sr_0)
        mfcc_0_list.append(mfcc_0)
        
    elif machine_train['FAN_TYPE'][i]==2:
        audio_path_2 = machine_train['SAMPLE_ID'][i]+'.wav'
        y_2, sr_2 = librosa.load('sound/train/'+audio_path_2)
        stft_2 = librosa.stft(y_2, n_fft = n_fft, hop_length = hop_length)
        spectrogram_2 = np.abs(stft_2) 
        mel_spec_2 = librosa.feature.melspectrogram(S=spectrogram_2, sr=sr_2, n_mels=n_mels, hop_length=hop_length)
        S_dB_2 = librosa.power_to_db(mel_spec_2, ref=np.max) 
        mfcc_2 = librosa.feature.mfcc(S=S_dB_2, sr=sr_2)
        mfcc_2_list.append(mfcc_2)
```
