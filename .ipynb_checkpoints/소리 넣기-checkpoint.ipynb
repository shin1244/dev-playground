```python
import librosa
import librosa.display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def to_mels(file_path):
    
    y, sr = librosa.load(file_path)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
    S_db = librosa.power_to_db(mel_spec, ref=np.max)
    
    return S_db 
```

```python
img = to_mels('D:/overlaysound/overlay1131.wav')
gray_image = np.expand_dims(img, axis=-1)
```

```python
img = to_mels('D:/overlaysound/overlay1131.wav')
gray_image = np.expand_dims(img, axis=-1)
librosa.display.specshow(gray_image, hop_length=512, x_axis='time', y_axis='mel', fmax=8000)
plt.show()
```
