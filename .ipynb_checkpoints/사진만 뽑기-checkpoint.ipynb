```python
import os
import glob
import shutil

# 이미지 파일 확장자 목록
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']

# 대상 폴더 경로
source_folder = 'C:/Users/jg999/Desktop/Drunk.v5-zibs.voc/train'

# 저장 폴더 경로
target_folder = 'C:/Users/jg999/Desktop/Drunk.v5-zibs.voc/train1'

# 폴더 내의 파일 목록 조회
files = glob.glob(os.path.join(source_folder, '*'))

# 이미지 파일 필터링
image_files = [file for file in files if os.path.splitext(file)[1].lower() in image_extensions]

# 이미지 파일 저장
for image_file in image_files:
    # 저장할 파일 경로 생성
    target_file = os.path.join(target_folder, os.path.basename(image_file))
    
    # 이미지 파일 복사
    shutil.copy(image_file, target_file)
    
    print(f"{image_file}을 {target_file}로 저장했습니다.")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2 as cv
import os
import numpy as np
import matplotlib.pyplot as plt

datagen = ImageDataGenerator(
    width_shift_range = 0.1,
    height_shift_range = 0.3,
    rotation_range = 30
)
cham = 'C:/Users/jg999/Desktop/test'
cham_list = os.listdir(cham)
cmcr = 'C:/Users/jg999/Desktop/1234'  # 저장할 폴더 경로
c=0
for i in cham_list:
  img_path = os.path.join(cham, i)
  image = np.expand_dims(plt.imread(img_path), 0)
  for _ in datagen.flow(image, batch_size=1,
               save_to_dir = cmcr,
               save_prefix='cmcr',
               save_format='png'):
    c+=1
    if c > 1:
      break
```
