```python
import pandas as pd
import numpy as np
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('house_train.csv')
```

```python
df = pd.get_dummies(df)
df
```

```python
df = df.fillna(df.mean())
```

```python
df_corr = df.corr()
```

```python
df_corr['SalePrice'].sort_values(ascending=False).head(7)[1:].index
```

```python
sns.pairplot(df[['SalePrice','OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF','1stFlrSF']])
plt.show()
```

```python
cols_train=['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF']
```

```python
X_train_pre=df[cols_train]
```

```python
y = df['SalePrice'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_pre, y, test_size=0.2)
```

```python
model = Sequential()

model.add(Dense(16,input_dim=6,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=20) # ì°¸ê¸°
modelpath="house.hdf5"
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=0, save_best_only=True)
```

```python
history = model.fit(X_train, y_train, validation_split=0.25, epochs=2000, batch_size=32, callbacks=[early_stopping_callback, checkpointer])
```

```python
real_prices =[]
pred_prices = []
X_num = []
```

```python
n_iter = 0
Y_prediction = model.predict(X_test).flatten()
for i in range(25):
    real = y_test[i]
    prediction = Y_prediction[i]
    print("ì‹¤ì œê°€ê²©: {:.2f}, ì˜ˆìƒê°€ê²©: {:.2f}".format(real, prediction))
    real_prices.append(real)
    pred_prices.append(prediction)
    n_iter = n_iter + 1
    X_num.append(n_iter)
```

```python
plt.plot(X_num, pred_prices, label='predicted price')
plt.plot(X_num, real_prices, label='real price')
plt.legend()
plt.show()
```
